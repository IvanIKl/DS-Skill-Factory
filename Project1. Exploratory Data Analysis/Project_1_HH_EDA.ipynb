1). Начнем с простого - с признака **"Образование и ВУЗ"**. Его текущий формат это: **<Уровень образования год выпуска ВУЗ специальность...>**. Например:
* Высшее образование 2016 Московский авиационный институт (национальный исследовательский университет)...
* Неоконченное высшее образование 2000  Балтийская государственная академия рыбопромыслового флота…
Нас будет интересовать только уровень образования.

Создайте с помощью функции-преобразования новый признак **"Образование"**, который должен иметь 4 категории: "высшее", "неоконченное высшее", "среднее специальное" и "среднее".

Выполните преобразование, ответьте на контрольные вопросы и удалите признак "Образование и ВУЗ".

Совет: обратите внимание на структуру текста в столбце **"Образование и ВУЗ"**. Гарантируется, что текущий уровень образования соискателя всегда находится в первых 2ух слов и начинается с заглавной буквы. Воспользуйтесь этим.

*Совет: проверяйте полученные категории, например, с помощью метода unique()*


```python
def get_education(data_str):
    '''
    функция ищет и возвращает признак уровня образования  
    аргументы: 
        data_str - тип string. Строка, в которой происходит поиск признака образования
    возвращаемое значение: 
        search_str - тип string. Образование в соответствии с заданным шаблоном из списка. 
        not_found_str - тип string. Если соответствие не найдено, то 'не найдено'
    '''   
    # задаем шаблон поиска 
    # важен порядок: сначала ищем составные признаки образования из нескольких слов, потом из одного слова
    search_str_lst = ['неоконченное высшее','среднее специальное','высшее','среднее'] 
    not_found_str = 'не найдено'
    
    data_str_lst = data_str.lower().split()[:2]
    where_search = ''.join(data_str_lst)
    
    for search_str in search_str_lst:
        if search_str.replace(' ','') in where_search:
            return search_str
      
    return not_found_str


df['Образование'] = df['Образование и ВУЗ'].apply(get_education)
display(df['Образование'].unique())
display(df['Образование'].value_counts())

# удалаяем ненужный признак
df.drop(['Образование и ВУЗ'], axis=1, inplace=True)
```

3). Следующим этапом преобразуем признак **"Опыт работы"**. Его текущий формат - это: **<Опыт работы: n лет m месяцев, периоды работы в различных компаниях…>**. 

Из столбца нам необходимо выделить общий опыт работы соискателя в месяцах, новый признак назовем "Опыт работы (месяц)"

Для начала обсудим условия решения задачи:
* Во-первых, в данном признаке есть пропуски. Условимся, что если мы встречаем пропуск, оставляем его как есть (функция-преобразование возвращает NaN)
* Во-вторых, в данном признаке есть скрытые пропуски. Для некоторых соискателей в столбце стоит значения "Не указано". Их тоже обозначим как NaN (функция-преобразование возвращает NaN)
* В-третьих, нас не интересует информация, которая описывается после указания опыта работы (периоды работы в различных компаниях)
* В-четвертых, у нас есть проблема: опыт работы может быть представлен только в годах или только месяцах. Например, можно встретить следующие варианты:
    * Опыт работы 3 года 2 месяца…

В результате преобразования у вас должен получиться столбец, содержащий информацию о том, сколько месяцев проработал соискатель.
Выполните преобразование, ответьте на контрольные вопросы и удалите столбец **"Опыт работы"** из таблицы.


```python
def get_experience(data_str):
    '''
    функция ищет и возвращает признак длительности опыта работы  
    аргументы: 
        data_str - тип string. Cтрока, в которой происходит поиск 
    возвращаемое значение: 
        experience - тип integer. Опыт работы в месяцах
        если признак не указан/пропущен, то возвражается np.NaN 
        если признак не найден, то возвращается -1      
    ''' 
    if pd.isnull(data_str) or data_str.lower() == 'не указано':
        return np.nan
    
    year_lst = ['лет', 'год', 'года']
    month_lst = ['месяц', 'месяца', 'месяцев']
    
    data_lst = data_str.split()
             
    experience = -1
    if data_lst[3] in year_lst:
        experience = int(data_lst[2]) * 12
        if data_lst[5] in month_lst:
            experience += int(data_lst[4])
    elif data_lst[3] in month_lst:
        experience = int(data_lst[2])
            
    return experience

df['Опыт работы (месяц)'] = df['Опыт работы'].apply(get_experience)
# Проверка, что все нашлось
print (f"Не нашлось {df[df['Опыт работы (месяц)'] == -1].shape[0]} строк")
print (f"Медианный опыт работы {df['Опыт работы (месяц)'].median()} месяцев")   

# удалаяем ненужный признак
df.drop(['Опыт работы'], axis=1, inplace=True)       
        
    
```

    Еще один важный факт: при выгрузки данных у некоторых соискателей "потерялась" информация о готовности к командировкам. Давайте по умолчанию будем считать, что такие соискатели не готовы к командировкам.
    
Выполните преобразования и удалите столбец **"Город, переезд, командировки"** из таблицы.

*Совет: обратите внимание на то, что структура текста может меняться в зависимости от указания ближайшего метро. Учите это, если будете использовать порядок слов в своей программе.*


```python
def get_city_move_trip(data_str):
    '''
    функция ищет и возвращает признаки город проживания, готовность к переезду и готовность к командировкам 
    аргументы: 
        data_str - тип string. Cтрока, в которой происходит поиск 
    возвращаемое значение: 
        серия, содержащая 
            city город проживания (тип string),
            ready_to_move готовность к переезду (тип boolean),
            ready_to_trip готовность к командировкам (тип boolean) 
            found_move_block признак, что был найден блок текста о переезде (тип boolean)
            found_trip_block призник, что был найден блок текста о командировках (тип boolean)     
    ''' 
    
    data_lst = data_str.split(',')
    
    # Признак города
    million_cities = [
        'Новосибирск', 'Екатеринбург','Нижний Новгород','Казань',
        'Челябинск','Омск', 'Самара', 'Ростов-на-Дону', 'Уфа', 'Красноярск', 'Пермь',
        'Воронеж','Волгоград'
            ]        
    city_dict = {
        'Москва': 'Москва',
        'Санкт-Петербург': 'Санкт-Петербург'
    }    
    for million_city in million_cities:
        city_dict[million_city] = 'город-миллионник'        
    city = city_dict.get(data_lst[0].split()[0], 'другие')
        
    # Признак готовности к переезду
    # ищем блок текста о готовности к переезду
    search_str = 'перее'
    found_move_block = False
    ready_to_move = False
    for item in data_lst[1:]:
        item_lst = item.split()
        for word in item_lst:
            # проверка на наличие ключевого слова о переезде
            if word.find(search_str) != -1:
                found_move_block = True
                if 'не' not in item_lst:
                    ready_to_move = True
                break
    
    # Признак готовности к командировкам
    # ищем блок текста о готовности к командировкам
    search_str = 'командиров'
    found_trip_block = False
    ready_to_trip = False
    for item in data_lst[1:]:
        item_lst = item.split()
        for word in item_lst:
            # проверка на наличие ключевого слова о командировке
            if word.find(search_str) != -1:
                found_trip_block = True
                if 'не' not in item_lst:
                    ready_to_trip = True
                break
        
    return pd.Series([city, ready_to_move, ready_to_trip, found_move_block, found_trip_block])
    
df[['Город', 'Готовность к переезду', 'Готовность к командировкам', 'move_check', 'trip_check']] = \
df['Город, переезд, командировки'].apply(get_city_move_trip)

print(f"Не найдено строк с текстом о переезде {df['move_check'].count() - df['move_check'].sum()}")
print(f"Не найдено строк с текстом о командировке {df['trip_check'].count() - df['trip_check'].sum()}")

display(df['Город'].value_counts(normalize=True))

ready_to_all = df['Готовность к переезду'] & df['Готовность к командировкам']
print(f"Готовы к переезду и к командировкам {ready_to_all.mean():.2f}")

# удалаяем ненужные признаки
df.drop(['Город, переезд, командировки', 'move_check', 'trip_check'], axis=1, inplace=True)  

```

5). Рассмотрим поближе признаки **"Занятость"** и **"График"**. Сейчас признаки представляют собой набор категорий желаемой занятости (полная занятость, частичная занятость, проектная работа, волонтерство, стажировка) и желаемого графика работы (полный день, сменный график, гибкий график, удаленная работа, вахтовый метод).
На сайте hh.ru соискатель может указывать различные комбинации данных категорий, например:
* полная занятость, частичная занятость
* частичная занятость, проектная работа, волонтерство
* полный день, удаленная работа
* вахтовый метод, гибкий график, удаленная работа, полная занятость

Такой вариант признаков имеет множество различных комбинаций, а значит множество уникальных значений, что мешает анализу. Нужно это исправить!

Давайте создадим признаки-мигалки для каждой категории: если категория присутствует в списке желаемых соискателем, то в столбце на месте строки рассматриваемого соискателя ставится True, иначе - False.

Такой метод преобразования категориальных признаков называется One Hot Encoding и его схема представлена на рисунке ниже:
<img src=https://raw.githubusercontent.com/AndreyRysistov/DatasetsForPandas/main/ohe.jpg>
Выполните данное преобразование для признаков "Занятость" и "График", ответьте на контрольные вопросы, после чего удалите их из таблицы

```python
# формируем множество видов занятости
employment_lst = list(map(lambda x: x.split(','), list(df['Занятость'].unique())))
employment_lst = sum(employment_lst,[])
employment_lst = list(map(lambda x: x.strip(), employment_lst))
employment_set = set(employment_lst)

# формируем множество графиков работы
schedule_lst = list(map(lambda x: x.split(','), list(df['График'].unique())))
schedule_lst = sum(schedule_lst,[])
schedule_lst = list(map(lambda x: x.strip(), schedule_lst))
schedule_set = set(schedule_lst)

# формируем признаки
for employment in employment_lst:
    df[employment] = df['Занятость'].apply(lambda x: True if employment in x else False)
    
for schedule in schedule_lst:
    df[schedule] = df['График'].apply(lambda x: True if schedule in x else False)

project_volunt = df['проектная работа'] & df['волонтерство']
print(f"Ищут проектную работу или волонтёрство {project_volunt.sum():.0f} человек")

free_schedule = df['вахтовый метод'] & df['гибкий график']
print(f"Хотят работать вахтовым методом или с гибким графиком {free_schedule.sum():.0f} человек")

# удалаяем ненужные признаки
df.drop(['Занятость', 'График'], axis=1, inplace=True) 
```

Осталось только понять, откуда брать дату, по которой определяется курс? А вот же она - в признаке **"Обновление резюме"**, в нем содержится дата и время, когда соискатель выложил текущий вариант своего резюме. Нас интересует только дата, по ней бы и будем сопоставлять курсы валют.

Теперь у нас есть вся необходимая информация для того, чтобы создать признак "ЗП (руб)" - заработная плата в рублях.

После ответа на контрольные вопросы удалите исходный столбец заработной платы "ЗП" и все промежуточные столбцы, если вы их создавали.

Итак, давайте обсудим возможный алгоритм преобразования: 
1. Перевести признак "Обновление резюме" из таблицы с резюме в формат datetime и достать из него дату. В тот же формат привести признак "date" из таблицы с валютами.
2. Выделить из столбца "ЗП" сумму желаемой заработной платы и наименование валюты, в которой она исчисляется. Наименование валюты перевести в стандарт ISO согласно с таблицей выше.
3. Присоединить к таблице с резюме таблицу с курсами по столбцам с датой и названием валюты (подумайте, какой тип объединения надо выбрать, чтобы в таблице с резюме сохранились данные о заработной плате, изначально представленной в рублях). Значение close для рубля заполнить единицей 1 (курс рубля самого к себе)
4. Умножить сумму желаемой заработной платы на присоединенный курс валюты (close) и разделить на пропорцию (обратите внимание на пропуски после объединения в этих столбцах), результат занести в новый столбец "ЗП (руб)".


```python
usecols=['currency', 'date', 'close', 'proportion']
df_exchange = pd.read_csv('data/ExchangeRates.csv', usecols=usecols)

# Преобразовывыем даты
df['Обновление резюме'] = pd.to_datetime(df['Обновление резюме'], dayfirst=True)
df['Обновление резюме'] = df['Обновление резюме'].dt.date
df_exchange['date'] = pd.to_datetime(df_exchange['date'], dayfirst=True).dt.date


# Выделяем сумму и валюту
df['salary'] = df['ЗП'].apply(lambda x: float(x.split()[0]))
df['currency'] = df['ЗП'].apply(lambda x: x.split()[1])

cur_dict = {
    'грн.': 'UAH',
    'бел.руб.' : 'BYN',
    'сум': 'UZS'
}
df['currency'] = df['currency'].apply(lambda x: cur_dict.get(x,x))

# Объединяем базу резюме с таблицей курсов
df_merged = df.merge(df_exchange, how='left', left_on=['currency', 'Обновление резюме'], right_on=['currency', 'date'])

# Добавляем признаки close и proportion для рубля (единицы)
filt = df_merged[df_merged['currency'] == 'руб.']
df_merged.loc[filt.index, 'close'] = 1
df_merged.loc[filt.index, 'proportion'] = 1

# Считаем зарплату в рублях
df_merged['ЗП (руб)'] = df_merged['salary'] * df_merged['close'] / df_merged['proportion']

print(f"Медианная заработная плата {df_merged['ЗП (руб)'].median():.0f} руб.")

# удалаяем ненужные признаки
df_merged.drop(usecols, axis=1, inplace=True)
df_merged.drop(['ЗП', 'salary'], axis=1, inplace=True)
```

Попробуйте построить распределение признака в **логарифмическом масштабе**. Добавьте к графику линии, отображающие **среднее и границы интервала метода трех сигм**. Напомним, сделать это можно с помощью метода axvline. Например, для построение линии среднего будет иметь вид:

`histplot.axvline(log_age.mean(), color='k', lw=2)`

В какую сторону асимметрично логарифмическое распределение? Напишите об этом в комментарии к графику.
Найдите выбросы с помощью метода z-отклонения и удалите их из данных, используйте логарифмический масштаб. Давайте сделаем послабление на **1 сигму** (возьмите 4 сигмы) в **правую сторону**.

Выведите таблицу с полученными выбросами и оцените, с каким возрастом соискатели попадают под категорию выбросов?

```python
age = np.log(df_filter['Возраст'])
m = age.mean()
sigma = age.std()
left_border = m - 3*sigma
right_border = m + 3*sigma

# строим график
fig = plt.figure(figsize=(15,7))

hist = sns.histplot(
    data=age,
    bins=30    
)

hist.axvline(m, color='k', lw=2)
hist.axvline(left_border, color='k', lw=2)
hist.axvline(right_border, color='k', lw=2)
hist.set_title('Распределение признака возраст в логарифмическом масштабе', pad=20)

skew = age.skew()
if skew > 0:
    print (f'Правосторонняя ассиметрия {skew:.2f}')
elif skew < 0:
    print (f'Левосторонняя ассиметрия {skew:.2f}')
else:
    print('Распределение симметричное')
    
sns.despine()

# ищем выбросы
mask1 = age > m + 4*sigma
mask2 = age < m - 3*sigma
outliers = age[mask1 | mask2]

print(f'Количество удаленных выбросов {outliers.shape[0]}')
display(df_filter.loc[outliers.index])

# итоговый датафрейм
df_filtered = df_filter.loc[~df_filter.index.isin(outliers.index)]
print(f'Осталось {df_filtered.shape[0]} строк')
```

**Выводы:**  
Распределение возраста в логарифмическом масштабе имеет правостороннюю ассиметрию (0.45). По методу z-отклонений с условием 3 сигмы влево и 4 вправо выделяется 3 строчки с возрастами 15 и 100 лет. После их удаления в финальном наборе данных остается 44 482 строки.
